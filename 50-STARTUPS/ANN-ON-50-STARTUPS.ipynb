```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
startup = pd.read_csv('/content/drive/MyDrive/ANN/50_Startups (2).csv')
```

```python
startup = startup.drop(['State'],axis=1)
```

```python
train, test = train_test_split(startup, test_size=0.2, random_state=42)
```

```python
x_train = train.iloc[:,0:3].values.astype('float64')
```

```python
x_test = test.iloc[:,0:3].values.astype('float64')
```

```python
y_train = train.Profit.values.astype('float64')
```

```python
y_test = test.Profit.values.astype('float64')
```

```python
def norm_func(i):
  x = (i-i.min())/(i.max()-i.min())
  return(x)
```

```python
x_train = norm_func(x_train)
```

```python
x_test = norm_func(x_test)
```

```python
y_train = norm_func(y_train)
```

```python
y_test = norm_func(y_test)
```

```python
y_train = np_utils.to_categorical(y_train)
```

```python
y_test = np_utils.to_categorical(y_test)
```

```python
num_of_classes = y_test.shape[1]
```

```python
def design_mlp():
    model = Sequential()
    model.add(Dense(500,input_dim=3, activation="relu"))
    model.add(Dense(200, activation = "tanh"))
    model.add(Dense(100, activation = "tanh"))
    model.add(Dense(50, activation = "tanh"))
    model.add(Dense(num_of_classes , activation = "linear"))
    model.compile(loss = "mean_squared_error", optimizer ='adam', metrics=['accuracy'])
    return(model)
```

```python
model = design_mlp()
```

```python
model.fit(x=x_train, y=y_train, batch_size = 100, epochs=20)
```

```python
eval_score_test = model.evaluate(x_test, y_test ,verbose=1)
```

```python
print("Accuracy: %.3f%%" %(eval_score_test[1]*100))
```
